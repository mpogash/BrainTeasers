```python
# == Distirbutions Functions =================================
# import libraries 
import numpy as np
from scipy.optimize import least_squares

# 1D Normal Distribution
def normal_dist_1D(x_array, coeffs):
    mean = coeffs[0]
    std = coeffs[1]
    return 1/(std*np.sqrt(2*np.pi))*np.exp(-(x_array-mean)**2/(2*std**2))

# 1D Normal Distribution Minimization 
def min_fun_normal(coeff,x_array,y_array):
    return normal_dist_1D(x_array, coeff) - y_array
```

```python
# == Test Distributions Function ==============================
import numpy as np
import matplotlib.pyplot as plt
#from distributions import normal_dist_1D

mean_val = 23.8
std_val = 3
noise_level = 0.1
x_array = np.arange(-100,100)
```

```python
dist_normal_noise = noise_level*np.random.rand(len(x_array))
dist_normal = normal_dist_1D(x_array, [mean_val, std_val]) + dist_normal_noise
plt.plot(x_array, dist_normal)
```

```python

init_guess = [23, 4]
fit_normal = least_squares(min_fun_normal, init_guess, args=(x_array,dist_normal))

dist_normal_mean = fit_normal.x[0]
dist_normal_std = fit_normal.x[1]
print(f"normal distribution mean and standard deviation are {dist_normal_mean} , {dist_normal_std}")
```
